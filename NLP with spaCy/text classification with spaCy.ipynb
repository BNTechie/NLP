```python
from spacy.matcher import Matcher
```

```python
# Initialize the Matcher with the shared vocabulary
matcher = Matcher(nlp.vocab)

# Define a pattern to match adjectives followed by nouns (e.g., "great movie")
adj_noun_pattern = [{"POS": "ADJ"}, {"POS": "NOUN"}]
matcher.add("ADJ_NOUN_PATTERN", [adj_noun_pattern])

# Define a pattern to match monetary amounts (e.g., "$100 million")
money_pattern = [{"IS_CURRENCY": True}, {"LIKE_NUM": True}, {"LOWER": {"IN": ["million", "billion", "trillion"]}}]
matcher.add("MONEY_PATTERN", [money_pattern])
```

```python
# Process a sample of text from the dataset
for i in range(7):  # Let's process the first 7 reviews
    text = dataset['train'][i]['text']
    doc = nlp(text)
    
    # Apply the matcher to the document
    matches = matcher(doc)
    
    print(f"Review {i + 1}:")
    for match_id, start, end in matches:
        matched_span = doc[start:end]
        print(f"  Matched span: {matched_span.text}")
    print("\n")
```

# Preprocess data

```python
import random
import spacy
from spacy.util import minibatch


# Split the dataset
train_texts = dataset['train']['text']
train_labels = [{'cats': {'POSITIVE': label == 1, 'NEGATIVE': label == 0}} for label in dataset['train']['label']]
test_texts = dataset['test']['text']
test_labels = [{'cats': {'POSITIVE': label == 1, 'NEGATIVE': label == 0}} for label in dataset['test']['label']]

# Check lengths
print(f"Training texts: {len(train_texts)}, Training labels: {len(train_labels)}")
print(f"Test texts: {len(test_texts)}, Test labels: {len(test_labels)}")
```

# Create the text classification model

# Train the model

```python
from spacy.training.example import Example

# Create a blank English model
nlp = spacy.blank("en")

# Add the text classifier to the pipeline
textcat = nlp.add_pipe("textcat")

# Add labels to the text classifier
textcat.add_label("POSITIVE")
textcat.add_label("NEGATIVE")

# Initialize the model weights
nlp.initialize()
```

```python
# Prepare training data
train_data = list(zip(train_texts, train_labels))

# Training the model
n_iter = 3  # Number of training iterations

for i in range(n_iter):
    losses = {}
    # Shuffle the training data
    random.shuffle(train_data)
    for text, labels in train_data:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, labels)
        nlp.update([example], losses=losses)
    print(f"Iteration {i + 1} - Loss: {losses['textcat']}")
```

```python
# Evaluate the model
correct = 0
for i, text in enumerate(test_texts):
    doc = nlp(text)
    if doc.cats['POSITIVE'] > doc.cats['NEGATIVE']:
        pred_label = "POSITIVE"
    else:
        pred_label = "NEGATIVE"
    
    true_label = "POSITIVE" if test_labels[i]['cats']['POSITIVE'] else "NEGATIVE"
    
    if pred_label == true_label:
        correct += 1

accuracy = correct / len(test_texts)
print(f"Accuracy: {accuracy:.3f}")

# Make a prediction on new text
doc = nlp("This movie was amazing!")
print(doc.cats)
```
